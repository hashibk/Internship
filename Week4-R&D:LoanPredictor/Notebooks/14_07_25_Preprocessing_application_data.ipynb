```python
df.drop(df.columns[41:87], axis=1, inplace=True)
```

```python
df.head()
```

```python
le = LabelEncoder()

for item in nomimal_cols:
  df[item] = le.fit_transform(df[item])

weekday_mapper = {
    'MONDAY': 5,
    'TUESDAY': 7,
    'WEDNESDAY': 6,
    'THURSDAY': 4,
    'FRIDAY': 3,
    'SATURDAY': 2,
    'SUNDAY': 1
}

df['WEEKDAY_APPR_PROCESS_START'] = df['WEEKDAY_APPR_PROCESS_START'].map(weekday_mapper)
```

```python
df.dtypes
```

```python
df_new = df.copy() #copied a new dataframe here i will classify the original column 'NAME_TYPE_SUITE' into two groups, single and multiple
# Define mapping function
def classify_suite(value):
    if value in ['Unaccompanied', np.nan]:  # Single
        return 'Single'
    else:  # All others are considered Multiple
        return 'Multiple'

# Apply classification to the column
df_new['NAME_TYPE_SUITE'] = df_new['NAME_TYPE_SUITE'].apply(classify_suite)

df_new['NAME_TYPE_SUITE'] = le.fit_transform(df_new['NAME_TYPE_SUITE'])
```
